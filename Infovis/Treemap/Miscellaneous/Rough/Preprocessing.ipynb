```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("world_bank_development_indicators.csv")
df.describe()
```

```python
df.head()
```

```python
country_df = df.groupby(['country']).sum().reset_index()
country_df.set_index("country", inplace=True)
country_df.drop(['date'], axis=1, inplace=True)
country_df
```

```python
country_df.head()
```

```python
yearly_df = df.groupby('date').sum(numeric_only=True).reset_index()
yearly_df.set_index('date', inplace=True)
# yearly_df.drop(['F1'], axis=1, inplace=True)
yearly_df
```

```python
yearly_df.to_csv("year-wise.csv")
country_df.to_csv("country-wise.csv")
```
